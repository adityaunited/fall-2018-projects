```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
race4= pd.read_csv('2014.csv', na_values=['---','0'])
race4.head()
```

```python
race4.describe()
```

```python
race4.keys()
```

```python
reg='overall_time~bike_time+swim_time+run_time+age'
model=smf.ols(formula= reg, data=race4)
model.fit()
print(model.fit().summary())
```
